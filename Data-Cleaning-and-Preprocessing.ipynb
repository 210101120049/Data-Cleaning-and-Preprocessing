```python
# decide whether to do outliers tratment or not 
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
wisker(df['GDP'])
```

```python
for i in ['GDP','Total expenditure','Thinness 1-19 years','Thinness 5-9 years']
   lw,uw=wisker(df[i])

df[i]=np.where(df[i]<lw,lw,df[i])
df[i]=np.where(df[i]>lw,lw,df[i])
```
